```python
def connect(ur):
    uClient = urlopen(ur)
    page_html = uClient.read()

    uClient.close()
    
    page_soup = soup(page_html, "html.parser")

    container = page_soup.find_all("div",{"class": "_2kHMtA"})
    return container
```

```python
def dataframe(containers):
    data = {"Product_name":[],"Sale price":[],"Actual Price":[],"Ratting":[],"Person Ratting":[],"Offer":[]}
    for i in range(0,15):
        con = connect(containers[i])
        #Nazwa produktu
        for container in con:
            try:
                product_name = container.div.img["alt"]
                data["Product_name"].append(product_name)
            except:
                print("Nie odnaleziono nazwy produktu.")
                data["Product_name"].append("Nie odnaleziono nazwy")
            
            #Cena sprzedaży produktu
            s_price = container.findAll("div",{"class":"_30jeq3 _1_WHN1"})
            s_p = s_price[0].text
            data["Sale price"].append(s_p)
            
            #Rzeczywista cena produktu
            try:
                actual_price = container.findAll("div",{"class":"_3I9_wc _27UcVY"})
                a_p = actual_price[0].text
                data["Actual Price"].append(a_p)
            except IndexError:
                print("Cena jest taka sama jak cena sprzedaży")
                data["Actual Price"].append("Cena jest taka sama jak cena sprzedaży")
            
            #Ratting
            try:
                ratting = container.findAll("div",{"class":"_3LWZlK"})
                r = ratting[0].text
                data["Ratting"].append(r)
            except IndexError:
                data["Ratting"].append("Brak rattingu")
            
            #Ilość wystawionych ratingów
            try:
                p_retting = container.findAll("span",{"class":"_2_R_DZ"})
                p_r = p_retting[0].text
                data["Person Ratting"].append(p_r)
            except IndexError:
                data["Person Ratting"].append("Brak rattingu")
                        
            
            #Zniżka na produkt
            try:
                offer = container.findAll("div",{"class":"_3Ay6Sb"})
                off = offer[0].text
                data["Offer"].append(off)
            except IndexError:
                data["Offer"].append("Obecnie brak zniżki")
            
        print("done",i)
        
    return data
```

```python
dataset = dataframe(urls)
```

```python
df = pd.DataFrame(dataset)
df.dtypes
```

```python
df["Ratting"] = pd.to_numeric(df["Ratting"])
```

```python
df["No of Reviews"] = df["No of Reviews"].str.split(" ").str[0]
```

```python
df = df.drop(["Person Ratting","No of Reviews"],axis=1)
```
